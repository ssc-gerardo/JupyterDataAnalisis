```python
import pandas as pd
import numpy as np
from IPython.display import HTML
from io import BytesIO
import os
```

```python
file="20191126 Cargas en Sell Out.xlsx"
xl= pd.ExcelFile(file)
sheetnames=xl.sheet_names
file=[pd.read_excel(file, sheet_names=s) for s in sheetnames]

cargas=pd.DataFrame(file[0])    
cargas
```

```python
df= pd.crosstab(cargas.X_CLIENTE, [cargas.nIdPeriodo,cargas.D_FECHA], margins=True)
df.to_excel("fechas_titulo.xlsx", sheet_name="periodos" )
df
```
