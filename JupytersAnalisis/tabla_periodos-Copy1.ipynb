```python
import chardet
def find_encoding(fname):
    r_file = open(fname, 'rb').read()
    result = chardet.detect(r_file)
    charenc = result['encoding']
    return charenc

my_encoding = find_encoding('faltantes_cargas.txt')

lee = pd.read_csv('faltantes_cargas.txt', delimiter='!' , encoding=my_encoding)
lee
```

```python
df_2=lee.style.applymap(lambda x:'background-color: % s' % 'red' if x==0 else 'background-color:% s'%'green'if x==1 else 'background-color:% s'%'none',)
df_2
#y = lambda symbol: 'X' if symbol==True else 'O' if symbol==False else ' '
```
