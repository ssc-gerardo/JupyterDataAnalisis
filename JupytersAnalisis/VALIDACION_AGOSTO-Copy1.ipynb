```python
base_id_cliente = base_agosto.groupby('N_ID_CLIENTE')
base_p= list()
base_v= list()
for N_ID_CLIENTE, data_cliente in base_id_cliente:
    suma_p=data_cliente['F_PAQUETES'].sum()
    suma_v=data_cliente['F_VENTA_NETA'].sum()
    base_p.append(suma_p)
    base_v.append(suma_v)
    print("N_ID_CLIENTE:",N_ID_CLIENTE)
    print("",suma_p)
    print("",suma_v)
    print("\n")
```

```python
base_upc = base_agosto.groupby('X_SKU')
base_upc_p= list()
base_upc_v= list()
for X_SKU, data_upc in base_upc:
    suma_p=data_upc['F_PAQUETES'].sum()
    suma_v=data_upc['F_VENTA_NETA'].sum()
    base_p.append(suma_p)
    base_v.append(suma_v)
    print("X_SKU:",X_SKU)
    print("",suma_p)
    print("",suma_v)
    print("\n")
```

```python
import chardet
def find_encoding(fname):
    r_file = open(fname, 'rb').read()
    result = chardet.detect(r_file)
    charenc = result['encoding']
    return charenc

my_encoding = find_encoding('carga agosto/al super/al_super.csv')

carga_al_super = pd.read_csv('carga agosto/al super/al_super.csv', delimiter=',', encoding=my_encoding)

carga_al_super
```
